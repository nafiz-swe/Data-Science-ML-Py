We will use **tips**, **iris** and another very famous dataset **titanic** for these exercises. These datasets are included in the seaborn's built-in dataset library.<br>

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Do the following tasks:** 
* set style to '`white`'
* load titanic dataset into '`titanic`', tips into '`tips`' and iris into '`iris`'

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

from IPython.display import display

# Set style
sns.set_style('white')

# Load datasets
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')

titanic_formatted = titanic.copy() 
titanic_formatted['age'] = titanic_formatted['age'].map('{:.1f}'.format) 
titanic_formatted['fare'] = titanic_formatted['fare'].map('{:.4f}'.format)

display(titanic_formatted.head().style.set_table_styles(
    [{'selector': 'table', 'props': [('border', '2px solid black')]}]))
```

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

tips_formatted = tips.copy()
tips_formatted['total_bill'] = tips_formatted['total_bill'].map('{:.2f}'.format)
tips_formatted['tip'] = tips_formatted['tip'].map('{:.2f}'.format)

display(tips_formatted.loc[:, ['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']].head().style.set_table_styles(
    [{'selector': 'table', 'props': [('border', '2px solid black')]}]))
```

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

iris_formatted = iris.copy()
iris_formatted['sepal_length'] = iris_formatted['sepal_length'].map('{:.1f}'.format)
iris_formatted['sepal_width'] = iris_formatted['sepal_width'].map('{:.1f}'.format)
iris_formatted['petal_length'] = iris_formatted['petal_length'].map('{:.1f}'.format)
iris_formatted['petal_width'] = iris_formatted['petal_width'].map('{:.1f}'.format)

display(iris_formatted.loc[:, ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']].head().style.set_table_styles(
    [{'selector': 'table', 'props': [('border', '2px solid black')]}]))

```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

tips = sns.load_dataset("tips")

plt.figure(figsize=(4, 3))
ax = sns.boxplot(x="day", y="tip", hue="sex", data=tips, palette={"Female": "green", "Male": "#558ED5"})

plt.xlabel("day", fontsize=8)
plt.ylabel("tip", fontsize=8)
print("<matplotlib.axes._subplots.AxesSubplot at 0x11c2002e8>")

ax.yaxis.grid(False)

for spine in ['top', 'right', 'bottom', 'left']:
    plt.gca().spines[spine].set_color('gray')
    plt.gca().spines[spine].set_linewidth(1.5)
    plt.gca().spines[spine].set_linestyle('-')

plt.xticks(fontsize=8)

plt.legend(title='sex', loc='upper right', fontsize='xx-small', title_fontsize='x-small', frameon=False)

plt.show()
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

tips = sns.load_dataset("tips")

print("<seaborn.axisgrid.JointGrid at 0x1fe2e00aec0>")

g = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', color='#558ED5', height=4.5)

g.set_axis_labels('total_bill', 'tip', fontsize=8)

for spine in g.ax_joint.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(0.6)
```

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

titanic = sns.load_dataset("titanic")
print("<matplotlib.axes._subplots.AxesSubplot at 0x11c657b00>")

plt.figure(figsize=(4, 3))
sns.histplot(titanic['fare'], kde=False, color='#2AB842', bins=30, edgecolor='none') 

plt.xlabel('fare', fontsize=8)
plt.ylabel('') 

# Remove internal gridlines
plt.gca().yaxis.grid(False)
plt.gca().xaxis.grid(False)

for spine in plt.gca().spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(0.8)

# Show the plot
plt.show()
```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
## Code here and leave the next cell empty, so that you don't lose the plot!

titanic = sns.load_dataset("titanic")
print("<matplotlib.axes._subplots.AxesSubplot at 0x11c787320>")

plt.figure(figsize=(4, 3))
sns.boxplot(x='class', y='age', hue='survived', data=titanic, palette={0: '#558ED5', 1: '#FFA07A'})

plt.xlabel('class', fontsize=8)
plt.ylabel('age', fontsize=8)

for spine in ['top', 'right', 'bottom', 'left']:
    plt.gca().spines[spine].set_color('black')
    plt.gca().spines[spine].set_linewidth(0.8)

plt.gca().yaxis.grid(False)
plt.gca().xaxis.grid(False)

plt.xticks(fontsize=7)

plt.legend(title='alive', loc='upper right', fontsize='xx-small', title_fontsize='x-small', 
           labels=['No', 'Yes'], frameon=False)

plt.yticks(range(0, 81, 10))

# Show the plot
plt.show()

```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

titanic = sns.load_dataset("titanic")

plt.figure(figsize=(4, 3))
g = sns.jointplot(x='fare', y='age', data=titanic, kind='scatter', color='#558ED5', height=4.5, marginal_kws=dict(bins=30, fill=True))

# Set the labels
g.set_axis_labels('fare', 'age', fontsize=8)

for spine in g.ax_joint.spines.values():
    spine.set_visible(False)  # Remove all borders

for ax in [g.ax_marg_x, g.ax_marg_y]:
    ax.yaxis.grid(False)
    ax.xaxis.grid(False)
    for spine in ax.spines.values():
        spine.set_visible(False)  
    ax.set_facecolor('white') 

g.ax_joint.set_facecolor('white') 

# Show the plot
plt.show()


```

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
print("<matplotlib.axes._subplots.AxesSubplot at 0x11cd02eb8>")


plt.figure(figsize=(4, 3))
sns.boxplot(x='class', y='age', hue='class', data=titanic, palette='Set3', dodge=False)
plt.legend([], [], frameon=False)

plt.xlabel('class', fontsize=8)
plt.ylabel('age', fontsize=8)

plt.gca().spines['top'].set_color('black')
plt.gca().spines['right'].set_color('black')
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.gca().spines['top'].set_linewidth(1.5)
plt.gca().spines['right'].set_linewidth(1.5)
plt.gca().spines['bottom'].set_linewidth(1.5)
plt.gca().spines['left'].set_linewidth(1.5)

plt.gca().yaxis.grid(False)
plt.gca().xaxis.grid(False)

plt.yticks(range(0, 81, 10))

plt.show()
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
titanic = sns.load_dataset("titanic")
print("<matplotlib.axes._subplots.AxesSubplot at 0x11d25c668>")

plt.figure(figsize=(5, 3))
sns.countplot(x="sex", hue="class", data=titanic, palette={"First": "blue", "Second": "green", "Third": "red"})

plt.xlabel("Sex", fontsize=8)
plt.ylabel("Count", fontsize=8)

plt.legend(title='Class', frameon=False, fontsize='small', title_fontsize='small')

plt.gca().yaxis.grid(False)
plt.gca().xaxis.grid(False)

plt.show()
```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

print("<matplotlib.text.Text at 0x11eb448d0>")


# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Select only numerical columns for correlation calculation
numeric_cols = titanic.select_dtypes(include=['float64', 'int64']).columns
corr = titanic[numeric_cols].corr()

# Create a heatmap
plt.figure(figsize=(4, 3))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.show()

```
