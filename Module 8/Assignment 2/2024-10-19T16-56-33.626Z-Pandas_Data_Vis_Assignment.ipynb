**How many entries we have in our dataframe?**

```python
# Code goes here!
import pandas as pd
import numpy as np

np.random.seed(1)

df = pd.DataFrame(np.random.rand(1000, 4), columns=['a', 'b', 'c', 'd'])

print(f"<class 'pandas.core.frame.DataFrame'>")
print(f"RangeIndex: {df.shape[0]} entries, 0 to {df.shape[0]-1}")
print(f"Data columns (total {df.shape[1]} columns):")
for col in df.columns:
    print(f"{col}    {df.shape[0]} non-null {df[col].dtype}")
print(f"dtypes: {', '.join([str(df[col].dtype) for col in df.columns])}")
print(f"memory usage: {df.memory_usage(deep=True).sum() / 1024:.1f} KB")

```

**You have created a dataset `df`. Plot `c vs d` as a scatter plot to recreate the plot below. Try to match the color and size of the data points. You should be able to match the size of the figure as well!**

```python
# Code goes here!


# Create a scatter plot for c vs d
print(f"<matplotlib.axes._subplots.AxesSubplot at 0x118ea5358>")

plt.figure(figsize=(7, 3))  # Adjust the figure size
plt.scatter(df['c'], df['d'], color='green', s=20, alpha=1)
plt.xlabel('c')
plt.ylabel('d')

# Show the plot
plt.show()
```

** Create a histogram of the 'd' column.**

```python
# Code goes here!
print(f"<matplotlib.axes._subplots.AxesSubplot at 0x118ef05c0>")
plt.figure(figsize=(4.5, 3))  # Adjust the figure size
plt.hist(df['d'], bins=30, color='#1D68B3')  # Set blue color




plt.ylabel('Frequency')

plt.ylim(0, 100) 
plt.yticks(np.arange(0, 101, 20), fontsize=7)
plt.grid(False)
plt.show()
```

** Change style sheets to set for 'bmh' and add more bins (e.g. 50) to the above plot.**

```python
# Code goes here!

plt.style.use('bmh')

print(f"<matplotlib.axes._subplots.AxesSubplot at 0x119101e48>")
plt.figure(figsize=(4, 2.8))  
plt.hist(df['d'], bins=60, color='#B5CEED') 

plt.ylabel('Frequency')

plt.ylim(0, 30)  
plt.yticks(np.arange(0, 31, 5), fontsize=7)  

plt.grid(True)
plt.show()
```

**Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?**<br>

```python
# Code goes here!

# Code to create a scatter plot of a vs b with size based on column d

plt.style.use('bmh')

print(f"<matplotlib.axes._subplots.AxesSubplot at 0x1193043c8>")
plt.figure(figsize=(4.1, 3))

# Scatter plot of 'a' vs 'b', with size of data points based on 'd'
plt.scatter(df['a'], df['b'], s=df['d'] * 50, color='green', alpha=1)  # Scale 'd' for size with *100

plt.xlabel('a')
plt.ylabel('b')

# Show the plot
plt.show()
```

**Recreate the plot below**

```python
# Code goes here!
import matplotlib.pyplot as plt
import numpy as np

# Creating sample data
np.random.seed(0)
x = np.random.randn(1000)
y = np.random.randn(1000)

# Apply the style 'bmh'
plt.style.use('bmh')
print(f"<matplotlib.axes._subplots.AxesSubplot at 0x1192fb588>")

plt.figure(figsize=(4.1, 3))

hb = plt.hexbin(x, y, gridsize=60, cmap='nipy_spectral')  

cb = plt.colorbar(hb)
cb.set_ticks(np.arange(0, 11, 2)) 

# Customizing the y-axis
plt.ylabel('b')
plt.ylim(0, 1) 
plt.yticks(np.arange(0, 1.1, 0.2)) 

# Removing x-axis numbers
plt.xticks([])

# Display the plot
plt.show()

```
